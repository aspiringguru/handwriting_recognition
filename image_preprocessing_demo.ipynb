```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
print ("libraries loaded")
```

```python
image = Image.open("./images/digit_four.png")
print(type(image))
plt.imshow(image)
```

```python
image_L = image.convert('L')
print(type(image_L))
print ("np.array(image_L).shape = ", np.array(image_L).shape)
plt.imshow(image_L)
#does weird things to colours, does yield an X by X array.
# cf above with X by X by 3 array when using convert('LA')
image_L_nparray = np.array(image_L)
print ("# of pixels = ", image_L_nparray.shape[0]*image_L_nparray.shape[1])
print ("len(image_L_nparray.nonzero()) = ", len(image_L_nparray.nonzero()[0]))
#NB: we have zero nonzero values!!! cf images from MNIST dataset with majority values = 0
print (image_L_nparray[0:1][0:5])
print (image_L_nparray[0:1][0:5].max())
print (image_L_nparray[0:1][0:5].min())
image_L.save('image_L','png')
```

```python
image_l28 = image_L.resize((28,28))
print (type(image_l28))
print (np.array(image_l28).shape)
print (np.array(image_l28)[0][0:28])
print (np.array(image_l28)[0][0:28].max())
print (np.array(image_l28)[0][0:28].min())
print (np.array(image_l28)[14][0:28])
print (np.array(image_l28)[14][0:28].max())
print (np.array(image_l28)[14][0:28].min())
plt.imshow(image_l28)
image_l28.save('image_l28','png')
```

```python
def binarize_array(numpy_array, threshold=200):
    """Binarize a numpy array."""
    for i in range(len(numpy_array)):
        for j in range(len(numpy_array[0])):
            if numpy_array[i][j] > threshold:
                numpy_array[i][j] = 255
            else:
                numpy_array[i][j] = 0
    return numpy_array
```

```python
image_threshold = binarize_array( np.array(image_l28), 100)
print (type(image_threshold))
print ("image_threshold.shape = ", image_threshold.shape)
print (image_threshold[0][0:28])
print (image_threshold[14][0:28])
#convert numpy.ndarray back to image
plt.imshow(image_threshold, cmap='gray')
#convert numpy.ndarray to PIL image then write to file
#image_threshold.save('image_threshold','png')
```
